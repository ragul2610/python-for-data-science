## Task 2

```python
def seperate_data_types(df):
    categorical = []
    numeric = []
    
    for column in df.columns:
        if(df[column].nunique())<30:
            categorical.append(column)
        else:
            numeric.append(column)
    return categorical,numeric
```

```python
categorical, numeric = seperate_data_types(df)
```

```python
categorical
```

```python
numeric
```

```python
df.info()
```

```python
print(tabulate({"Categorical":categorical,"Numeric":numeric},headers=['categorical','numeric']))
```

Conclusion - Seperated the categorical and numerical columns from the data set.

## Task 3 

```python
def info_of_cat(col): 
    print(f"Unique values in {col} are: {df[col].unique()}") # unique: returns the unique values in the col
    print(f"Mode of {col} is {df[col].mode()[0]}")  # mode: returns the mode of the column
    print(f"Number of missing values in {col} is {df[col].isnull().sum()}") # isnull().sum() counts the number of null in dataframe
      
```

## Task 5 

```python
def info_of_numeric(col):
    print(f"The mean of column {col} is {df[col].mean()}")
    print(f"The median of column {col} is {df[col].median()}")
    print(f"The mode of column {col} is {df[col].mode()[0]}")
    print(f"The std of column {col} is {df[col].std()}")
    print(f"Number of missing values in column {col} is {df[col].isnull().sum()}")
```

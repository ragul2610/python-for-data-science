## Task 1 

```python
import pandas as pd
```

```python
data = pd.read_csv("DS1_C4_S4_Test_Scores_Data_Practice.csv")
```

```python
data
```

```python
data.isnull().sum()
```

```python
data["gender"].value_counts()
```

```python
data["gender"]=data["gender"].fillna("Male")
```

```python
a=data["test1"].mean()
a
```

```python
data["test1"] = data["test1"].fillna(a)
```

```python
b = data["test2"].median()
b
```

```python
data["test2"] =data["test2"].fillna(b)
```

```python
data.isnull().sum()
```

Conclusion - Loaded the data and it is found that the null values in the data are being replaced by mean,median and mode.

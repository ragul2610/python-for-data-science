## Task 3 

```python
attrited_rarely = df[(df['Attrition'] == 'Yes') & (df['BusinessTravel'] == 'Travel_Rarely')]
```

```python
job_role_counts = attrited_rarely['JobRole'].value_counts()
```

```python
plt.figure(figsize=(10, 6))

job_role_counts.plot(kind='barh', color='skyblue')

plt.xlabel('Count of Attrited Employees')
plt.ylabel('Job Role')
plt.title('Job Role-wise Count of Attrited Employees (Traveling Rarely)')

plt.show()
```

# Conclusion - It is found that More number of attrited employees are at the job role as Laboratory Technician.

## Task 4 

```python
# Filter for employees with backgrounds in life science and medical education
filtered_df = df[df['EducationField'].isin(['Life Sciences', 'Medical'])]
```

```python
# Group the filtered data by "JobRole" and count the attrition
attrition_counts = filtered_df.groupby('JobRole')['Attrition'].value_counts().unstack().fillna(0)
```

```python
plt.figure(figsize=(10, 6)) 

attrition_counts.plot(kind='barh', stacked=True, color=['green', 'red'], alpha=0.8)

plt.xlabel('Count of Employees')
plt.ylabel('Job Role')
plt.title('Job Role-wise Employee Attrition (Life Sciences & Medical Education Background)')

plt.show()
```

# Conclusion - This the job role wise attited and not attrited for employee only in life science and medical field.

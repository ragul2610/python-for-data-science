### Task 6

```python
df.City_Mileage_km_litre.isnull().sum()
```

```python
df.Highway_Mileage_km_litre.isnull().sum()
```

```python
m=df.City_Mileage_km_litre.isnull().mean()
m
```

```python
m1=df.Highway_Mileage_km_litre.isnull().mean()
m1
```

```python
df.City_Mileage_km_litre.fillna(m,inplace=True)
```

```python
df.Highway_Mileage_km_litre.fillna(m1,inplace=True)
```

```python
f=df.groupby(by=["Make"])["City_Mileage_km_litre"].mean()
f=f.sort_values()
f
```

```python
plt.figure(figsize=(10,7))
plt.barh(f.index,f.values)
plt.show()
```

```python
f2=df.groupby(by=["Make"])["Highway_Mileage_km_litre"].mean()
f2=f2.sort_values()
f2
```

```python
plt.figure(figsize=(10,7))
plt.barh(f2.index,f2.values)
plt.show()
```

Conclusion -The null values in city and highway mileage column are treated and  the mean value for both city and highway mileage is found for all the manufacturers and shown in Horizontal bar chart .

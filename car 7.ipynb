```python
import pandas as pd
import matplotlib.pyplot as plt
import statistics as st
```

```python
car_df = pd.read_csv("DS1_C4_S7_Car_Data_Challenge.csv")
```

```python
car_df
```

```python
car_df['Kerb_Weight'] = car_df['Kerb_Weight'].str.replace(r"kg",'')
car_df
```

```python
data  = car_df[['Make','Cylinders','Displacement','Kerb_Weight','Fuel_Tank_Capacity_litre','City_Mileage_km_litre',
                      'Highway_Mileage_km_litre','Power','Torque']]
```

```python
data
```

```python
data.isnull().sum()
```

```python
data.dropna(subset=['Make','Cylinders','Displacement','Kerb_Weight','Fuel_Tank_Capacity_litre','City_Mileage_km_litre',
                      'Highway_Mileage_km_litre','Power','Torque'],inplace=True)
```

```python
data
```

IConclusion - The data is loaded and then drop the missing values for the specific columns mentioned in task.

### Task 2 

```python
data['Kerb_Weight'] = data['Kerb_Weight'].str.replace(r"kg",'')
data
```

```python
data["Torque"] = data["Torque"].str.strip()  
data['Torque'] = data['Torque'].str.replace(r"-",'', regex=True) 
data['Torque'] = data['Torque'].str.replace(r"[a-zA-Z]*@[0-9]*[a-zA-Z]*",'', regex=True) 


data["Power"] = data["Power"].str.strip () 
data['Power'] = data['Power'].str.replace(r"[a-zA-Z]*@[0-9]*[a-zA-Z]*",'', regex=True) 
```

```python
data
```

Interpretation - Replaced the Torque with only numeric values , Power and KG value with " " .

```python
data['Kerb_Weight'] = pd.to_numeric(data['Kerb_Weight'], errors='coerce')
data['Power'] = pd.to_numeric(data['Power'], errors='coerce')
data['Torque'] = pd.to_numeric(data['Torque'], errors='coerce')
```

### Task 4 

```python
manufacturers = ['Renault', 'Toyota', 'Honda']
filtered_data = data[data['Make'].isin(manufacturers)]

correlation= filtered_data.corr()
correlation
```

```python
city_mile = 'City_Mileage_km_litre'
corr_city = abs(correlation[city_mile])
corr_city = corr_city.sort_values(ascending=False)
corr_city[:5]
```

```python
fig, ax = plt.subplots(1,3, figsize=(8,4), sharey=True)

ax[0].scatter(x=filtered_data['Power'], y=filtered_data['City_Mileage_km_litre'])
ax[1].scatter(x=filtered_data['Kerb_Weight'], y=filtered_data['City_Mileage_km_litre'],color='green')
ax[2].scatter(x=filtered_data['Displacement'], y=filtered_data['City_Mileage_km_litre'],color='red')
plt.show()
```

```python
highway_mile = 'Highway_Mileage_km_litre'
corr_high = abs(correlation[highway_mile])
corr_high = corr_high.sort_values(ascending=False)
corr_high[:5]
```

```python
fig, ax = plt.subplots(1,3, figsize=(8,4), sharey=True)

ax[0].scatter(x=filtered_data['Power'], y=filtered_data['Highway_Mileage_km_litre'])
ax[1].scatter(x=filtered_data['Kerb_Weight'], y=filtered_data['Highway_Mileage_km_litre'],color='green')
ax[2].scatter(x=filtered_data['Displacement'], y=filtered_data['Highway_Mileage_km_litre'],color='red')
plt.show()
```

Conclusion -  Both City and Highway mileage the column of power, kerb weight, displacement are highest in y axis and reduces, so they have a negative correlation with City mileage of cars.

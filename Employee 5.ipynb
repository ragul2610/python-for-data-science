# Employee Data Analysis

## Task 1 

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('DS1_C4_S5_Employee_Data_Practice.xlsx',sheet_name=0)
```

```python
data
```

```python
plt.figure(figsize=(8,6))
colors = ['blue']
plt.bar(data["Department"],data["Annual Salary ($)"],color=colors)
plt.xlabel("Department")
plt.ylabel("Annual Salary ($)")
plt.title("department wise salary")
plt.show()
```

Conclusion - Total Annual sales of employee in each departments is represented as a bar graph.

## Task 2 

```python
x=data.groupby(by=['Department','Gender'])['Gender'].count()
x
```

```python
df=pd.DataFrame(x)
```

```python
df
```

```python
x.plot(x='Department',kind='bar',stacked=False,title='department wise gender',xlabel='Department',ylabel='Gender')
plt.show()
```

Conclusion -These are the department wise count of Male and Female employees is represented as a bar graph

## Task 4 

```python
filtered=data.groupby(by=['Department'])['Annual Salary ($)'].sum()
filtered
```

```python
pd.DataFrame(filtered)
```

```python
filtered.plot(kind="pie")
plt.pie(filtered,autopct='%.2f%%')
plt.title("Sum of salary in each depatment")
plt.show()
```

Conclusion - These are the department wise ratio for the sum of salaries for all employees in each department is represented in a pie chart

## Task 5  

```python
data.sort_values("Age",ascending = False)
```

```python
plt.hist(data["Age"],bins=5)
plt.xlabel("Age")
plt.ylabel("Employee_Code")
plt.title("Age wise count of employees")
plt.show()
```

Conclusion - This is the distribution of employee count by age group in a histogram.

## Task 6 

```python
plt.figure(figsize=(15, 8))
plt.scatter(data['Work_Experience'], data['Annual Salary ($)'], color='g', marker='x')
plt.title('Salary Trend based on Employee Experience')
plt.xlabel('Work Experience (years)')
plt.ylabel('Annual Salary ($)')
plt.grid(True)
```

Conclusion - This shows the Salary trend based on the employee experience in a Scatter plot.

## Task 7 

```python
plt.figure(figsize=(10, 6))
plt.boxplot([data[data['Department'] == 'IT ']['Annual Salary ($)'],
             data[data['Department'] == 'HR']['Annual Salary ($)'],
             data[data['Department'] == 'Finance']['Annual Salary ($)'],
             data[data['Department'] == 'Sales']['Annual Salary ($)']],
            labels=['IT', 'HR', 'Finance', 'Sales'])
plt.title('Department-wise Salary Spread')
plt.xlabel('Department')
plt.ylabel('Annual Salary ($)')
plt.grid(True)
plt.show()
```

Conclusion - These are the Department wise Spread of employees salary by using box plot

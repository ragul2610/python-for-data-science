# Used cars:Sales Data Analysis 

## Task 1

```python
import pandas as pd
data = pd.read_csv('DS1_C4_S4_Used_Car_Prices_Data_Challenge.csv')
```

```python
data
```

```python
data.isnull().sum()
```

```python
data1=data.dropna()
data1
```

```python
data1.isnull().sum()
```

Conclusion - Loaded the data and dropped the rows with missing valuesand cleaned the data

## Task 2 

```python
tho = data1[(data1["year"] == 2014) & (data1["state"] == "ca")]
```

```python
tho
```

```python
sort_tho = tho.sort_values('mp', ascending=True)
sort_tho
```

```python
top_100=sort_tho.head(100)
top_100
```

Conclusion - The data is filtered for top 100 lowest priced cars from 2014 with CA registration.

## Task 3 

```python
pd.crosstab(sort_tho.make,sort_tho.mp)
```

```python
pd.crosstab(top_100.make,top_100.mp)
```

Conclusion - These are those top 100 lowest priced cars from 2014 with CA registration in a cross-tab

## Task 4 

```python
x = sort_tho[(sort_tho.make == 'Chevrolet')]
x
```

```python
new=x[['make','mp']]
```

```python
new
```

```python
discount=new[["mp"]]*0.9
discount
```

Conclusion - 10% discount is added for the chevrolet cars.these are the prices of chevrolet cars after discount

## Task 5 

```python
gruup = pd.pivot_table(top_100,index=['interior'],values=['mp'],aggfunc='mean')
```

```python
gruup
```

Conclusion - These are the average priced of the above top 100 cars based on interior colors

## Task 6 

```python
filtered_data = data1[(data1['year']>= 2010) & (data1['year']<=2015)]
```

```python
filtered_data
```

```python
a = pd.pivot_table(filtered_data,index=['make','color'],values=['mp'],aggfunc='count')
```

```python
a
```

Conclusion - These are the color wise availability of cars for each manufacturers from 2010 - 2015
